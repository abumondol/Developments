```python
import numpy as np
import pickle
```

```python
# Read Right hand data for StevenTech Free Living
file_path = "C:\\ASM\\DevData\\eating\\java\\text_data_for_matlab\\steven\\data_free\\steven_free_accel_";
free_data = [];

for subject in range(11):
    sess_count = 2
    if subject == 9:
        sess_count = 5
    
    subject_data = [];
    for sess in range(sess_count):        
        file_name = file_path+str(subject)+"_"+str(sess)+".csv";
        print(file_name)
        data = np.genfromtxt(file_name, delimiter=',')
        print(data.shape)
        subject_data.append(data)
        
    free_data.append(subject_data)

with open("C:\\ASM\\DevData\\eating\\python\\steven_free_data.pkl", 'wb') as handle:
    pickle.dump(free_data, handle, protocol = pickle.HIGHEST_PROTOCOL)
    
print("Reading done")            
        
```

```python
# Read Results for StevenTech Free Living
def read_steven_free_results(type):
    file_path = "C:\\ASM\\DevData\\eating\\java\\text_data_for_matlab\\steven\\results_free\\steven_res"+type+"_free_bites_"
    free_data = [];

    for subject in range(11):
        sess_count = 2
        if subject == 9:
            sess_count = 5

        subject_data = [];
        for sess in range(sess_count):        
            file_name = file_path+str(subject)+"_"+str(sess)+".csv";
            print(file_name)
            data = np.genfromtxt(file_name, delimiter=',')
            print(data.shape)
            subject_data.append(data)

        free_data.append(subject_data)

    with open("C:\\ASM\\DevData\\eating\\python\\steven_free_results"+type+".pkl", 'wb') as handle:
        pickle.dump(free_data, handle, protocol = pickle.HIGHEST_PROTOCOL)
        
    print("Reading done")            
```

```python
for type in ["0", "1", "2", "_left"]:
    read_steven_free_results(type)
```
