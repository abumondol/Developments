```python
import numpy as np
import pickle 
import matplotlib.pyplot as plt

import data_manager as dm
```

```python
def find_meals_proximity(mp, res):
    distance = 60*16 # 60 seconds    
    bites = mp[res[:, 1]==1, 0]
    #print(bites*0.0625)
    
    clusters = []
    c = [bites[0]]
    for i in range(1, len(bites)):
        if bites[i] - bites[i-1]<=distance:
            c.append(bites[i])
        else:            
            clusters.append(c)
            c = [bites[i]]               
    clusters.append(c)
    
    #res = []    
    #count = len(clusters)
    #i = 0
    #while i <count:
    #    if len(clusters[i]>=4):
            
            
        
    
    if len(clusters) ==0:
        return []
        
    meals = np.zeros((len(clusters), 3))
    for i in range(len(clusters)):        
        meals[i, 0] = clusters[i][0]
        meals[i, 1] = clusters[i][-1]
        meals[i, 2] = len(clusters[i])        
    return meals
```

```python
def evaluate_result(gt_meals, res_meals):
    print("Ground Truth:")
    a = np.copy(gt_meals)
    a[:, :2] = a[:, :2]*0.0625
    a = a.astype(int)    
    print(a)    
    
    print("Result:")
    if len(res_meals) == 0:
        print(res_meals)
    else:
        a = np.copy(res_meals)
        a[:, :2] = a[:, :2]*0.0625
        a = a.astype(int)
        print(a)    
```
