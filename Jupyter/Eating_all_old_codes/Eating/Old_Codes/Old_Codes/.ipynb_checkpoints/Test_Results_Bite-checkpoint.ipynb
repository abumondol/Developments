```python
########## Free Performance #################
path = "C:\\ASM\\DevData\\eating\\python\\steven_free_data.pkl"
with open(path, "rb") as file:
    data = pickle.load(file)        
```

```python
def find_clusters(bites, eps, minPts):    
    count = len(bites)
    print("Detected bite count: ", count)
    if count == 0:
        return []
    
    pt = np.zeros((count, 2))
    
    for i in range(count):
        ct = 0
        j = i-1
        while j>=0 and bites[i] - bites[j]<=eps:
            ct+=1
            j-=1
            
        j = i+1
        while j<count and bites[j] - bites[i]<=eps:
            ct+=1
            j+=1
        
        pt[i, 1] = ct
        if ct >= minPts:
            pt[i, 0] = 1
    
    clusters = []
    six = -1
    eix = -1
    for i in range(count):        
        if pt[i, 0] == 1 and six < 0:            
            j = i-1
            while j>=0 and bites[i] - bites[j]<=eps:                
                j-=1
            six = j+1
            
        elif pt[i, 0] != 1 and six >= 0:                        
            j = i+1
            while j<count and bites[j] - bites[i]<=eps:                
                j+=1
            eix = j-1
            clusters.append([six, eix])
            six = -1
            eix = -1
            
    return clusters
        
def find_meals(mp, res):
    eps = 1*60*16 # 1 minutes
    minPts = 3
    bites = mp[res[:, 1]==1, 0]
    #print(bites*0.0625)
    
    clusters = find_clusters(bites, eps, minPts)
    if len(clusters) ==0:
        return []
    
    clusters = np.array(clusters)
    meals = np.zeros((len(clusters), 3))
    for i in range(len(clusters)):
        six = clusters[i, 0]
        eix = clusters[i, 1]
        meals[i, 0] = bites[six]
        meals[i, 1] = bites[eix]
        meals[i, 2] = eix-six+1
    return meals
```

```python
def evaluate_result(gt_meals, res_meals):
    print("Ground Truth:")
    a = np.copy(gt_meals)
    a[:, :2] = a[:, :2]*0.0625
    a = a.astype(int)    
    print(a)    
    
    print("Result:")
    if len(res_meals) == 0:
        print(res_meals)
    else:
        a = np.copy(res_meals)
        a[:, :2] = a[:, :2]*0.0625
        a = a.astype(int)
        print(a)    
```

```python
for subject in range(11):    
    for sess in range(len(res[subject]["free"])):        
        print("\nSubject {}, Session {}>>".format(subject, sess))
        r = res[subject]["free"][sess]
        mp = mps[3][subject][sess]
        #print(len(data[subject+2][sess][1]))
        annot = data[subject][sess][1]
        
        mp_count = len(mp)
        annot_count = len(annot)
        res_count = len(r)        
        
        print("mp_count, res_count: {}, {}".format(mp_count, res_count))
        print("annot_count: {}".format(annot_count))        
        
        res_meals = find_meals(mp, r)
        #print(annot*0.0625)
        evaluate_result(annot, res_meals)
        
        flag = True
        if flag:
            continue
        
        fig = plt.figure(figsize=(14,10))
        subplot = fig.add_subplot(111)        
        subplot.scatter(mp[:, 0], r[:, 1], marker=".")
        #subplot.plot(d[:, 2], label='Y')
        #subplot.plot(d[:, 3], label='Z')
        #subplot.legend()
        
        
        for i in range(annot_count):            
            ix1 = annot[i, 0]
            ix2 = annot[i, 1]
            meal_type = annot[i, 2]
            
            subplot.plot([ix1, ix1], [-5, 5], '--', c = 'b')
            subplot.plot([ix2, ix2], [-5, 5], c = 'g')
        
        #plt.show()
        plt.savefig("C:\\ASM\\DevData\\eating\\python\\figs2\\"+str(subject)+"_"+str(sess)+".png")
        
```
