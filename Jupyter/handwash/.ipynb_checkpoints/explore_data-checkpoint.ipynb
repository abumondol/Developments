```python
import numpy as np
import pickle
import os
from collections import Counter
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
id_list = []
path = "C:/ASM/DevData/handwash/labelled_data/meta_data.pkl"
with open(path, "rb") as file:
    gt = pickle.load(file)
```

```python
print("Total days: ", len(gt))
nids = []
hw = []
data = []
for e in gt:
    nids.append(e[0])
    data.extend(e[4])
    h = e[3]    
    if len(hw) ==0:
        hw = h
    elif len(hw)>0:
        hw = np.concatenate((hw,h))

a = Counter(nids)
nids = list(set(nids))
print("Total subjects:", len(nids))
print(nids)
print(a)
```

```python
print(len(data))
data = np.array(data)
d = data[:, 1] - data[:, 0]
s = sum(d)

h = s//3600
m = (s - h*3600)//60
sec = s - h*3600 - m*60
print("Total durations: {} sec = {} hour, {} min, {} sec".format(s, h, m, sec) )
```

```python
print("HW count: ", len(hw))
durations = hw[:,1] - hw[:, 0]
print("average: ", np.mean(durations))
print("median: ", np.median(durations))
print("min: ", np.amin(durations))
print("max: ", np.amax(durations))

```

```python
plt.hist(durations,  bins=50)
```

```python
s = []
s2 = []
for e in gt:        
    if len(e[3])>0:
        s.append(e[0])
    if len(e[4])>0:
        s2.append(e[0])
        
s = list(set(s))
s2 = list(set(s2))
s.sort()
s2.sort()
print(len(s), ", ", len(s2))
print(s)
print(s2)
        
```

```python
for e in gt:
    h = e[2]
    h = h[h[:, 0]==39, 1:]
        
    for i in range(len(h)-1):
        if h[i, 1] > h[i+1, 0]:
            print("Problem :", e[0], ", ", e[1])
            print(h[i, :])
            print(h[i+1, :])
        elif h[i+1, 1] - h[i, 0]<70:
            print("Close :", e[0], ", ", e[1])
            print(h[i, :])
            print(h[i+1, :])
    
```

```python
full = 0
partial1 = 0
partial2 = 0
partial3 = 0
hw_count = 0
for e in gt:
    print(e[0], ", ", e[1])
    print(e[3])
    print(e[4])
    h = e[3]
    fcount = 0
    for i in range(len(h)):
        hw_count += 1
        st = h[i, 0]
        et = h[i, 1]
        seg_times = e[4]
        for j in range(len(seg_times)):            
            if st>=seg_times[j][0] and et<=seg_times[j][1]:
                fcount+=1
                full+=1
                print("   >> Full: {}-{} <-> {}-{}".format(h[i, 0], h[i, 1], seg_times[j][0], seg_times[j][1]))
                break
            elif st>=seg_times[j][0] and st<seg_times[j][1] and et>seg_times[j][1]:
                partial1 += 1
                print("   >> Partial 1: {}-{} <-> {}-{}".format(h[i, 0], h[i, 1], seg_times[j][0], seg_times[j][1]))
                break
            elif st<seg_times[j][0] and et>=seg_times[j][0] and et<=seg_times[j][1]:
                partial2 += 1
                print("   >> Partial 2: {}-{} <-> {}-{}".format(h[i, 0], h[i, 1], seg_times[j][0], seg_times[j][1]))
                break
            elif st<seg_times[j][0] and et>seg_times[j][1]:
                partial3 += 1
                print("   >> Partial 3: {}-{} <-> {}-{}".format(h[i, 0], h[i, 1], seg_times[j][0], seg_times[j][1]))
                break
        
    print("  ===========>>>>> full covered {}/{}".format(fcount, len(h)) )

no = hw_count - full - partial1 - partial2 - partial3
print("Total: ", hw_count)
print("Full covered: ", full)
print("Partial covered:{}, {}, {}".format(partial1, partial2, partial3))
print("No covered: ", no)

            
            
```
